# Employee Insights - SQL Data Analysis

## Pregunta 1: Cuantos empleados han tenido mas de un puesto

```sql
SELECT 
    emp_no,
    COUNT(DISTINCT title) AS num_titulos
FROM titles
GROUP BY emp_no
HAVING num_titulos > 1
ORDER BY num_titulos DESC;
```

### Resultado
Los empleados tienen hasta 3 titulos diferentes a lo largo de su carrera.

### Insight
La empresa promueve internamente a su gente, lo cual es positivo para la retencion de talento.

---

## Pregunta 2: Cuantos empleados han cambiado de departamento

```sql
SELECT 
    emp_no,
    COUNT(DISTINCT dept_no) AS num_departamentos
FROM dept_emp
GROUP BY emp_no
HAVING num_departamentos > 1
ORDER BY num_departamentos DESC;
```

### Resultado
Los empleados que se movieron cambiaron a un maximo de 2 departamentos.

### Insight
La movilidad interna es baja. Esto puede indicar que los empleados se especializan en sus areas o que hay pocas oportunidades de rotacion interdepartamental.

---

## Pregunta 3: Cual es la antiguedad promedio de los empleados

```sql
SELECT 
    ROUND(AVG(DATEDIFF(CURDATE(), hire_date) / 365), 1) AS antiguedad_promedio_anos
FROM employees;
```

### Resultado
36.2 anos de antiguedad promedio.

### Insight
La fuerza laboral es extremadamente estable y leal. Esto refleja un ambiente de trabajo que retiene a su personal a largo plazo.

---

## Pregunta 4: Que departamento tiene mas empleados actualmente

```sql
SELECT 
    d.dept_name,
    COUNT(de.emp_no) AS total_empleados
FROM dept_emp de
JOIN departments d ON de.dept_no = d.dept_no
WHERE de.to_date = '9999-01-01'
GROUP BY d.dept_name
ORDER BY total_empleados DESC;
```

### Resultado
| Departamento      | Total Empleados |
|-------------------|-----------------|
| Development       | 61,386          |
| Production        | 53,304          |
| Sales             | 37,701          |
| Customer Service  | 17,569          |
| Research          | 15,441          |
| Marketing         | 14,842          |
| Quality Management| 14,546          |
| Human Resources   | 12,898          |
| Finance           | 12,437          |

### Insight
La empresa esta orientada a tecnologia y produccion. Development y Production concentran mas del 50% de la fuerza laboral total.

---

## Pregunta 5: Cuantos empleados hay por genero

```sql
SELECT 
    gender,
    COUNT(*) AS total
FROM employees
GROUP BY gender;
```

### Resultado
| Genero | Total   |
|--------|---------|
| M      | 179,973 |
| F      | 120,051 |

### Insight
El 60% de los empleados son hombres y el 40% mujeres. Existe una brecha de genero del 20% que el area de HR deberia considerar atender con estrategias de inclusion y equidad.

---

## Resumen General

La empresa cuenta con una fuerza laboral de 300,024 empleados con una antiguedad promedio de 36.2 anos, lo que refleja alta estabilidad y retencion. Se observa promocion interna activa aunque la movilidad entre departamentos es limitada. El area de Development es la mas grande de la organizacion. Se recomienda revisar la brecha de genero y explorar programas de rotacion interdepartamental para fomentar el desarrollo profesional.
