## Conclusiones y Recomendaciones

1. **Competitividad salarial:** Con un promedio de ~$72K y un máximo de $158K, la empresa tiene una estructura salarial robusta. Se recomienda hacer benchmarking con la industria para confirmar competitividad.

2. **Equidad de género:** La brecha salarial es mínima (0.09%), un indicador muy positivo. Sin embargo, se recomienda trabajar en la representación femenina, que actualmente es del 40%.

3. **Equilibrio departamental:** La diferencia entre Sales ($89K) y Human Resources ($63K) es de ~28%. Se sugiere revisar si esta brecha está justificada por estructura de comisiones o si requiere ajuste.

4. **Gestión de altos ingresos:** El 22% de la nómina se concentra en empleados con salarios >$90K. Se recomienda establecer planes de retención y sucesión para este grupo estratégico.

5. **Revisión de salarios bajos:** Los 6,960 empleados que ganan menos de $50K representan un grupo a monitorear para reducir riesgo de rotación.
