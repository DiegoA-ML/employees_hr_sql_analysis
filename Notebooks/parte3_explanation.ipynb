# Parte 3: Desempeño Departamental — Reporte de Resultados

## Contexto de Negocio
Este análisis evalúa cómo están estructurados y distribuidos los departamentos de la empresa al cierre de 2002. Permite a la dirección y a RH tomar decisiones informadas sobre asignación de recursos, contratación, estabilidad de liderazgo y retención por área.

### 3.1 ¿Cómo se distribuye la fuerza laboral entre departamentos?

**Necesidad de Negocio:** Entender la distribución de headcount para apoyar decisiones de presupuesto y contratación por área.

```python
-- Pregunta 3.1
SELECT 
    d.dept_name AS Departamento,
    COUNT(de.emp_no) AS TotalEmpleados
FROM dept_emp de
JOIN departments d ON de.dept_no = d.dept_no
WHERE de.from_date <= '2002-12-31'
  AND de.to_date > '2002-12-31'
GROUP BY d.dept_name
ORDER BY TotalEmpleados DESC;
```

### Resultado

| Departamento       | TotalEmpleados |
|--------------------|----------------|
| Development        | 61,386         |
| Production         | 53,304         |
| Sales              | 37,701         |
| Customer Service   | 17,569         |
| Research           | 15,441         |
| Marketing          | 14,842         |
| Quality Management | 14,546         |
| Human Resources    | 12,898         |
| Finance            | 12,437         |

### Insight
**Development** es el departamento más grande con **61,386 empleados**, seguido por **Production** con **53,304**. Entre ambos concentran más del **50%** de la fuerza laboral activa. La empresa está claramente orientada hacia tecnología y producción. En contraste, **Finance** (12,437) y **Human Resources** (12,898) son los departamentos más pequeños, aunque críticos para la operación.

---

### 3.2 ¿Qué departamentos han tenido más cambios de gerente?

**Necesidad de Negocio:** Detectar inestabilidad en el liderazgo departamental que pueda impactar cultura y retención.

```python
-- Pregunta 3.2
SELECT 
    d.dept_name AS Departamento,
    COUNT(*) AS CambiosDeGerente
FROM dept_manager dm
JOIN departments d ON dm.dept_no = d.dept_no
GROUP BY d.dept_name
ORDER BY CambiosDeGerente DESC;
```

| Customer Service   | 4               |
| Quality Management | 4               |
| Production         | 4               |
| Development        | 2               |
| Marketing          | 2               |
| Research           | 2               |
| Human Resources    | 2               |
| Finance            | 2               |
| Sales              | 2               |

### Insight
**Customer Service**, **Quality Management** y **Production** tienen el doble de cambios de gerente que el resto de los departamentos. Esto puede ser señal de mayor exigencia operativa, mayor presión por resultados o menor estabilidad en estas áreas. Para RH, la continuidad del liderazgo es clave para la cultura departamental y la retención del talento; por ello se recomienda investigar las causas de esta rotación e implementar planes de sucesión específicos.

### 3.3 ¿Cuál es la antigüedad de los gerentes actuales en cada departamento?

**Necesidad de Negocio:** Identificar qué departamentos cuentan con liderazgo experimentado y cuáles con liderazgo más reciente.

```python
-- Pregunta 3.3
SELECT 
    d.dept_name AS Departamento,
    CONCAT(e.first_name, ' ', e.last_name) AS NombreGerente,
    dm.from_date AS InicioComoGerente,
    ROUND(DATEDIFF('2002-12-31', dm.from_date) / 365.25, 2) AS AnosComoGerente
FROM dept_manager dm
JOIN departments d ON dm.dept_no = d.dept_no
JOIN employees e ON dm.emp_no = e.emp_no
WHERE dm.to_date > '2002-12-31'
ORDER BY AnosComoGerente DESC;
```

### 3.4 ¿Qué departamentos tuvieron más contrataciones en los últimos 5 años (1998–2002)?

**Necesidad de Negocio:** Identificar dónde está ocurriendo el crecimiento más reciente para orientar el presupuesto de reclutamiento.

```python
-- Pregunta 3.4
SELECT 
    d.dept_name AS Departamento,
    COUNT(*) AS ContratacionesUltimos5Anos
FROM employees e
JOIN dept_emp de ON e.emp_no = de.emp_no
JOIN departments d ON de.dept_no = d.dept_no
WHERE e.hire_date BETWEEN '1998-01-01' AND '2002-12-31'
GROUP BY d.dept_name
ORDER BY ContratacionesUltimos5Anos DESC;
```

### Resultado

| Departamento       | Contrataciones (1998–2002) |
|--------------------|---------------------------|
| Development        | 1,588                     |
| Production         | 1,363                     |
| Sales              | 973                       |
| Customer Service   | 459                       |
| Marketing          | 415                       |
| Research           | 400                       |
| Quality Management | 379                       |
| Human Resources    | 340                       |
| Finance            | 331                       |

### Insight
**Development** y **Production** siguen siendo los motores de crecimiento en headcount también en el periodo reciente, reproduciendo la distribución general de la empresa. **Finance** y **Human Resources** presentan el menor crecimiento en contrataciones recientes; si la empresa aspira a mayor agilidad organizacional y soporte al crecimiento, podría considerar reforzar estas áreas de manera selectiva.

### 3.5 ¿Qué departamentos tienen la antigüedad promedio más baja?

**Necesidad de Negocio:** Identificar áreas que podrían tener menor retención para diseñar intervenciones focalizadas.

```python
-- Pregunta 3.5
SELECT 
    d.dept_name AS Departamento,
    COUNT(*) AS TotalEmpleados,
    ROUND(AVG(DATEDIFF('2002-12-31', e.hire_date) / 365.25), 2) AS AntiguedadPromedioAnios
FROM employees e
JOIN dept_emp de ON e.emp_no = de.emp_no
JOIN departments d ON de.dept_no = d.dept_no
WHERE de.from_date <= '2002-12-31'
  AND de.to_date > '2002-12-31'
GROUP BY d.dept_name
ORDER BY AntiguedadPromedioAnios ASC;
```

### Resultado

| Departamento       | Total Empleados | Antigüedad Promedio (años) |
|--------------------|-----------------|---------------------------|
| Marketing          | 14,842          | 12.90                     |
| Quality Management | 14,546          | 12.90                     |
| Sales              | 37,701          | 12.91                     |
| Human Resources    | 12,898          | 12.91                     |
| Development        | 61,386          | 12.92                     |
| Customer Service   | 17,569          | 12.92                     |
| Research           | 15,441          | 12.92                     |
| Production         | 53,304          | 12.93                     |
| Finance            | 12,437          | 12.93                     |

### Insight
La variación entre departamentos es prácticamente inexistente (apenas **0.03 años** de diferencia entre el más bajo y el más alto). La retención es uniformemente alta en toda la organización al cierre de 2002. Ningún departamento muestra señales claras de problemas de retención bajo esta métrica.

## Resumen General: Parte 3

La empresa presenta una estructura departamental altamente concentrada en tecnología y producción, con **Development** y **Production** dominando tanto en headcount actual como en contrataciones recientes.

Esto contrasta con áreas como **Finance** y **Sales**, donde el liderazgo es muy consolidado.
